# Handige functies in SymPy <a name="sympy"></a>

### De Heaviside functie

Laten we de eerste handige functie importeren: `Heaviside`. **Run de cel:** 

```python
from sympy import Heaviside
from sympy.abc import t # dit importeert t als een symbolische variabele
```

We kunnen nu een signaal definiëren met een functie, bijvoorbeeld
$f(t) = 
\begin{cases}
3 & \text{ voor }\ 1 \leq t<3 \\
0 & \text{ elders}
\end{cases}
$


**Run de cel:**

```python
def signaal1(t):
    return 3*(Heaviside(t-1)-Heaviside(t-3))
```

Het printen van deze functie levert geen mooie output op. Probeer maar.

```python
print(signaal1(t))
```

Het object `signaal1(t)` is een "callable" Python functie. Om de functie op een mooie en leesbare manier weer te geven, importeren we `sympify`. **Run de cel:**

```python
from sympy import sympify
sympify(signaal1(t))
```

De functie `sympify` moet ook worden toegepast bij het plotten van het signaal met de SymPy-plotfunctie. Voor het plotten importeren we eerste de module `matplotlib`, en stellen we de afmetingen van de plot in. **Run onderstaande cellen:**

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (5, 3)
```

```python
from sympy import plot
plot(sympify(signaal1(t)),(t,-1,5),xlim=[-1,5], ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plt.show()
```

### Een reeks verschoven kopieën van een functie maken
Voor begrip van Fourierreeksen is het handig om de grafiek van een signaal met de grafiek van zijn reekso te kunnen vergelijken.

Hieronder wordt de functie `verschoven_kopie` gedefinieerd waarmee met een signaal dat op slechts één periode gedefinieerd is, een nieuw signaal gemaakt kan worden dat bestaat uit een aantal verschoven kopieën van het originele signaal. Zo lijkt het nieuwe signaal periodiek en is vergelijking met de Fourierreeks eenvoudiger.

```python
def verschoven_kopie(f0, t0, T, N, p0):
    """
    Creëert een nieuw signaal bestaande uit verschoven kopieën van het inputsignaal.

    Parameters:
    - f0: het inputsignaal
    - t0: het vensterbegin
    - T : de vensterbreedte
    - N : het aantal kopieën
    - p0: de verschuiving van de kopieën p0*T naar rechts
    
    Returns:
    - Het nieuwe signaal.
    """

    t1 = t0 + T
    def signaal(x):
        kopie_signaal = 0
        for n in range(p0, p0+N):
            kopie_signaal += (Heaviside(t-n*T-t0)-Heaviside(t-n*T-t1))*f0(t-n*T)
        return kopie_signaal
    return signaal
```

Laten we de functie `verschoven_kopie` toepassen op `signaal1` om een nieuw signaal te maken. **Run onderstaande cel:**

```python
signaal2 = verschoven_kopie(signaal1,0,4,4,0)

plot(sympify(signaal2(t)),(t,-5,20),xlim=[-5,20], ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plt.show()
```

# De reële Fourierreeks berekenen

Met de SymPy functie `fourier_series` kan een deel van de reële Fourierreeks van een signaal berekend worden. Deze functie neemt als input het signaal, en een triplet met de variabele en het interval waarover de Fourierreeks bepaald moet worden.

```python
from sympy import fourier_series
```

De functie waarvan de reële Fourierreeks wordt berekend, moet periodiek zijn. Neem bijvoorbeeld de functie 
$f(t) = 
\begin{cases}
3 & \text{ voor }\ 1\leq t<3 \\
0 & \text{ elders}
\end{cases}
$
$\ \ \ $ en verder periodiek met periode 4.

Dit signaal kunnen we bouwen (op een eindig interval) met de functie `verschoven_kopie`.

```python
signaal4 = verschoven_kopie(signaal1,0,4,4,-1)

plot(sympify(signaal4(t)),(t,-4,12),xlim=[-4,12],ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plt.show()
```

De SymPy functie `fourier_series` neemt één periode van de periodieke functie als input. Willen we de reële Fourierreeks bepalen van `signaal4`, dan hoeven we alleen `signaal4` als input te geven met t van 0 tot 4. (Bijvoorbeeld; een ander interval dat één periode vangt werkt ook.) Het is wat betreft rekenkracht beter om `signaal1` als input te geven op dit interval (de kopieën zijn overbodig).

```python
fourierreeks1 = fourier_series(signaal1(t), (t, 0, 4))
sympify(fourierreeks1) # dit is beter leesbaar dan print()
```

De eerste termen van de reële Fourierreeks worden getoond. We kunnen aangeven voor hoeveel frequenties de reeks berekend dient te worden. Hiervoor gebruiken we de functie `truncate`. Bijvoorbeeld de eerste vier frequenties:

```python
fourierreeks1_n4 = fourierreeks1.truncate(4)
sympify(fourierreeks1_n4)
```

Laten we deze Fourierreeks plotten.

```python
plot(fourierreeks1_n4,(t,-4,12),xlim=[-4,12], ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plt.show()
```

En nu de eerste zes frequenties:

```python
fourierreeks1_n6 = fourierreeks1.truncate(6)
sympify(fourierreeks1_n6)
```

```python
plot(fourierreeks1_n6,(t,-4,12),xlim=[-4,12], ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plt.show()
```

Vergelijk de twee plots hierboven met de plot van het oorspronkelijke `signaal4`. Je ziet dat de Fourierreeks met de eerste zes frequenties het signaal (natuurlijk) beter benadert dan de Fourierreeks met de eerst vier frequenties. 

Hieronder worden de Fourierreeksen samen met het originele signaal geplot in één assenstelsel.

```python
plot(sympify(signaal4(t)),fourierreeks1_n4,(t,-4,12),xlim=[-4,12], ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plot(sympify(signaal4(t)),fourierreeks1_n6,(t,-4,12),xlim=[-4,12], ylim=[-1,4],xlabel='$t$',ylabel='$y$')
plt.show()
```
